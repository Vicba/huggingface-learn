# Synthetic data

Synthetic data is generated by a model to be used in place of real data or with real data. Here, by model, we donâ€™t mean only machine learning or deep learning models; they can be simple mathematical or statistical models too, like a set of (stochastic) differential equations modeling a physical or economic system.

Some of the non-exhaustive problems with real data are:
- It can be messy and very hard to deal with.
- Inter-company data sharing might not be possible due to privacy issues.
- Medical data is confidential and hence cannot be shared openly.
- It can be biased.
- Data collection and annotation can be expensive.

### challenges synthetic data

- Synthetic data is not inherently private: Synthetic data can also leak information about the data it was derived from and is vulnerable to privacy attacks. Significant care is required to generate private synthetic data.
- Outliers can be hard to capture privately: Outliers and low probability events, as are often found in real data, are particularly difficult to capture and to be privately included in a synthetic dataset.
- Empirically evaluating the privacy of a single dataset can be problematic: Rigorous notions of privacy (e.g., differential privacy) are a requirement on the mechanism that generated a synthetic dataset rather than on the dataset itself.
- Black box models can be particularly opaque when it comes to generating synthetic data: Overparameterised generative models excel in producing high-dimensional synthetic data, but the levels of accuracy and privacy of these datasets are hard to estimate and can vary significantly across produced data points.

# DCGAN

The text discusses the use of Deep Convolutional Generative Adversarial Networks (DCGAN) for generating synthetic medical images. A GAN consists of two neural networks, a Generator and a Discriminator, which compete to improve the quality of generated images. In the medical field, where data is often limited and sensitive, GANs have been used to generate synthetic images such as lung X-rays, retina images, brain scans, and liver images. Examples include generating brain PET images for Alzheimer's disease, synthetic liver images for CNN-based classification, and lung X-ray images to aid COVID-19 detection.

The text explains the use of DCGAN (Deep Convolutional Generative Adversarial Network) to generate synthetic lung images. DCGAN, introduced by Radford et al., utilizes convolutional layers to effectively capture spatial features. The Generator network transforms random noise into synthetic images, while the Discriminator distinguishes between real and synthetic images. Key components include using strided convolutions for downsampling instead of max-pooling.

The Generator's architecture involves a 100-dimensional noise input and outputs 128x128 images using convolutional layers, batch normalization, and activation functions like ReLU and Tanh. The Discriminator's architecture consists of convolutional layers with Leaky ReLU and Sigmoid activation, outputting the probability of an image being real or fake.

### challenges/opportunities using synthetic data
1. **Overfitting**: If synthetic data has overly consistent patterns (e.g., in color, size, background, or location), the model might overfit to those limited variations and perform poorly on real-world data.

2. **Bias in Synthetic Data**:
   - **Limited Diversity**: Synthetic data may lack the complexity found in real-world data, leading to inaccurate model predictions. For example, training on limited species data could result in poor performance when encountering other species.
   - **Copying Existing Biases**: If the original data used to generate synthetic data has biases, the model may inherit and perpetuate those biases.

3. **Computational Cost**: High-quality synthetic data generation can be computationally expensive. It should only be used if it ultimately saves time and resources.

4. **Quality of Synthetic Data**: Low-quality synthetic data can impair model performance. It's essential to evaluate the dataset using metrics like Frechet Inception Distance (FID), Inception Score (IS), or Classification Accuracy Score (CAS) to ensure that synthetic images resemble real data.

5. **Continuous Monitoring**: Even after training with synthetic data, models should be monitored in real-world scenarios to ensure they adapt to new patterns or trends not captured in the synthetic dataset.

In conclusion, synthetic data offers potential, but careful design and evaluation are necessary to avoid overfitting, biases, and quality issues, while ensuring the model performs well in real-world applications.
